```bash
#######################################
Input:
1. r1fetrev.fastq.gz (read 1 illumina female reads)
2. r2fetrev.fastq.gz (read 2 illumina female reads)

Function:
Gather the total number of reads, the average read basepair length and the standard read deviation. This is a prerequisite for
the configuration file of masurca.

Output:
1. Printed total of reads, average of read length and standard deviation of reads.
#######################################
```

```bash
#infer total number of reads/mean/SD required for Masurca config

fq=/workspace/hramzr/github/Trevally/PacBio/fastq/
echo "---forward---"
awk 'BEGIN { t=0.0;sq=0.0; n=0;} ;NR%4==2 {n++;L=length($0);t+=L;sq+=L*L;}END{m=t/n;printf("total %d avg=%f stddev=%f\n",n,m,sqrt(sq/n-m*m));}' ${fq}r1fetrev.fastq.gz
echo "---reverse---"
awk 'BEGIN { t=0.0;sq=0.0; n=0;} ;NR%4==2 {n++;L=length($0);t+=L;sq+=L*L;}END{m=t/n;printf("total %d avg=%f stddev=%f\n",n,m,sqrt(sq/n-m*m));}' ${fq}r2fetrev.fastq.gz
```

```bash
#######################################
Input:
1. String of text (Containing configuration settings for the CABOG assembler run)
2. String of text (Containing configuration settings for the FLYE assembler run)

Function:
Create the configuration files for the CABOG and FLYE assembler runs. This file directs to paired end forward and reverse read 
fastq files(r1fetrev.fastq.gz/r2fetrev.fastq.gz) and the PACBIO reads (femtrev.fastq.gz) 
and uses accompanying SD/mean read data. 

Output:
1. masurca_config (For CABOG run)
2. masurca_config (For FLYE run)
#######################################
```

```bash
#######################################
Input:
1. masurca_config (Configuration file for the masurca run with FLYE)

Function:
Run the assembler and create an assembly with the FLYE assembler.

Output:
1. Consensus.fasta (Assembled FASTA consensus file)
#######################################
```

```bash
#FLYE USED
module purge
module load perl/5.8.9
cd /workspace/hramzr/github/Trevally/PacBio/masurca/FLYE_RUN
logdir=/powerplant/workspace/hramzr/github/Trevally/PacBio/log/
mdir=/workspace/hramzr/github/Trevally/PacBio/masurca/MaSuRCA-3.3.7/bin/masurca
pwd
#run command
$mdir masurca_config
#start the assembly
bsub -m wkoppb37 -n 64 -J "FLYE" -o ${logdir}masurcaFLYE.out -e ${logdir}masurcaFLYE.err \
"./assemble.sh"
```
