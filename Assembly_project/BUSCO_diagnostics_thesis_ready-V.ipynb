```bash
#download teleost lineage data
wget https://busco-data.ezlab.org/v4/data/lineages/actinopterygii_odb10.2019-11-20.tar.gz
tar xzf actinopterygii_odb10.2019-11-20.tar.gz
```

```bash
#######################################
Input:
1. config(folder with augustus prerequisites from the software/bioinformatics path)

Function:
Copy the augustus-3.3.1 config recursively into the 'Diagnostics_assembly' folder working directory
in order to call a local AUGUSTUS_CONFIG_PATH 
later on, because of associated rights involved when running BUSCO related to augustus.
    
Output:
1. conig(folder with augustus prerequisites in the 'Diagnostics assembly' folder working directory)
#######################################
```

```bash
#config path dl to local since lacking rights for revision needed for busco
cp -r /software/bioinformatics/augustus-3.3.1/config /powerplant/workspace/hramzr/github/Trevally/PacBio/Diagnostics_assembly
```

```bash
#######################################
Input:
1. fetreval.fasta.contigs.fasta(FASTA format assembly of the canu_filtered run)

Function:
Call BUSCO stats which are needed for validation purposes of assemblies based on orthologues.
    
Output:
1. canu_filtered(folder with BUSCO output stats)
#######################################
```

```bash
#run busco and change august config path to be changed for lineage
# https://gitlab.com/ezlab/busco/issues/15
module purge
module load perl/bio-extras
module load augustus/3.2.2
module load pfr-python3/3.6.1
module load hmmer/3.1b2
module load blast/2.2.26 
module load BUSCO/v3.0.2     

logdir=/workspace/hramzr/github/Trevally/PacBio/log
genomecanu=/workspace/hramzr/github/Trevally/PacBio/canu_filtered/fetreval.fasta.contigs.fasta
export AUGUSTUS_CONFIG_PATH="/workspace/hramzr/github/Trevally/PacBio/Diagnostics_assembly/config"
bsub -o $logdir/busco.out -e $logdir/busco.err -n 8 -J "BUSCO" \
"run_BUSCO.py -r -m genome -i $genomecanu -o canu_filtered -l actinopterygii_odb10"
```

```bash
#######################################
Input:
1. short_summary_OUTPUT.txt(Text file with short summary output of the BUSCO run on the assembly)

Function:
Create an R script that plots the data from the short_summar_OUTPUT.txt file by running it in a python script called 
"generate_plot.py".
    
Output:
1. busco_figure.R (R script that can plot the short_summary_OUTPUT data)
#######################################
```

```bash
#create plot dir and generate R script to plot results 
module load pfr-python3/3.6.1
module load R/3.4.3
module load BUSCO/v3.0.2   
mkdir BUSCO_plots
cp short_summary_OUTPUT.txt BUSCO_plots/.

plt=/software/bioinformatics/BUSCO_v3.0.2/scripts/generate_plot.py
wkdir=/workspace/hramzr/github/Trevally/PacBio/Diagnostics_assembly/run_OUTPUT/BUSCO_plots/
python3 $plt -wd $wkdir  --no_r
```

```bash
#######################################
Input:
1. busco_figure.R (R script that generates a png from BUSCO output data)

Function:
Run the script 'busco_figure.R' to create a BUSCO data plot.

Output:
1. busco_figure.png (PNG file that contains a visualization of the short_summary_OUTPUT.txt file)
#######################################
```
