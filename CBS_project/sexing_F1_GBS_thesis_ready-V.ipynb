```bash
#######################################
Input:
1. Bam file names(TXT bash output)

Function:
Store filenames to run commands en masse and count the amount for a quick inventarisation.
    
Output:
1. bams.lst(TXT file with bam filenames per line)
2. Bash output of counted lines
#######################################
```

```bash
#Create list of all BAM files
ls | egrep "*.bam">bams.lst

# count bam_files
ls | egrep ".bam" | wc -l
```

```bash
#######################################
Input:
1. bams.lst(TXT file with bam filenames per line)

Function:
Index the bam files, which is a prerequisite to DEPTH calling.

Output:

1. *.bai files of every bam file that was in bam.lst
#######################################
```

```bash
while read -r BAM
do
    bsub \
    -o log/index.out -e log/index.err \
    "samtools index ${BAM}"
done < bams.lst
```

```bash
#######################################
Input:
1. bams.lst(TXT file with bam filenames per line)

Function:
Call depth of the sex gene subsection on the trevally001800 scaffold.

Output:
1. [bam file names]_depth (TXT files with depth per bam file present)
#######################################
```

```bash
module load samtools/1.7
while read -r BAM
do
    bsub \
    -o log/index.out -e log/index.err \
    "samtools depth -a -r trevally001800:200-1700 ${BAM} > depth_out/${BAM}_depth" 
done < bams.lst
```

```bash
#######################################
Input:
1. *_depth(TXT files with depth calculated per bam file)

Function:
Count the average depth in a depth_file and store this per depth file. 

Output:
1. average_depth(TXT file with average depth and associated name of bam file of origin)
#######################################
```

```bash
DO=depth_out/
rm average_depth
while read -r BAM
do
    cat ${DO}${BAM}_depth | awk '{ total += $3 } END {if (total == 0) print 0, '${BAM}'; else print total/NR,'${BAM}'}'>>average_depth
done < bams.lst
cat average_depth | awk '{print $1}'
```

```bash
#######################################
Input:
1. average_depth (TXT file with average_depth per bam file of origin)

Function:
Assign gender to specific depth_pattern and count genders after completion(near 50/50 is sign of correct calls)

Output:
1. average_depth_gendered(TXT file with gender assigned to name of bam file of origin per depth threshold)

Analysis conclusion:

Sex most likely is assigned correctly due to even number of calls. Chance of this being due to chance is LOW.
== gender most likely succesfully assigned
#######################################
```

```bash
cat average_depth | awk '{if($1<0.1) {print $0, "female"} else {print $0, "male"}}'>average_depth_gendered
cat average_depth_gendered | awk '{print $3}' | egrep "female" | wc -l
cat average_depth_gendered | awk '{print $3}' | egrep "^male" | wc -l
```
